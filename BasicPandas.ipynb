# Data Cleaning
* To identify null values in a Pandas dataframe, you can use several methods.

```python
data.isnull().any()
```

```python
data.isnull().sum()
```

* How to handle null values. Depending on how the null values are impacting your data you might have a different approach. We don't want to remove them if it means we might lose a record of our data!!

```python
# Filling null values with a specific value 

#df_filled = data.fillna(0)

# You can specify different fill values for different columns 

df_filled = data.fillna({'UPRN' : 0, 'Town': 'Unknown'})

print(df_filled.info())
```

```python
#Dropping columns with null values 

df_dropped_cols = data.dropna(axis=1)
(df_dropped_cols.info())


#Drop rows with null values in specific columns only 

df_cleaned = data.dropna(subset=['Town', 'UPRN'])
print(df_cleaned.info())
```
